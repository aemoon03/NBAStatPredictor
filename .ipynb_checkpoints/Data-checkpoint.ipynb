```python
import kagglehub 
import sqlite3 
import pandas as pd 
```

```python
path = kagglehub.dataset_download("wyattowalsh/basketball")

conn = sqlite3.connect(f'{path}/nba.sqlite')
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)
print(tables)
```

```python
table_name = tables['name'].iloc[0]  # Just an example
games = pd.read_sql(f"SELECT * FROM {table_name} WHERE game_date > '2003-01-01'", conn)
games.head()
```

```python
table_name = tables['name'].iloc[2]  # Just an example
other_stats = pd.read_sql(f"SELECT * FROM {table_name}", conn)
other_stats.head()
```

```python
table_name = tables['name'].iloc[6]  # Just an example
df = pd.read_sql(f"SELECT * FROM {table_name} WHERE game_date_est > '2003-01-01'", conn)
df.head()
```
